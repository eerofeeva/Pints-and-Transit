```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creating a bar chart for the top 50 start stations
station_count = final_df['start_station_name'].value_counts()
```

```python
station_df = final_df.loc[:, ["ride_id", "start_station_name", "end_station_name","start_lat", "start_lng", "end_lat", "end_lng"]]

starts_only_df = final_df.loc[:, ["ride_id", "start_station_name"]]
```

```python
start_stations = starts_only_df.groupby("start_station_name").count()
start_stations
```

```python
starts_sorted = start_stations.sort_values(by='ride_id', ascending=False)
top_50 = starts_sorted.head(50)
```

```python

top_50.plot(kind="bar", figsize=(20,3))
plt.title("Usage of Start Stations")
plt.ylabel("Number of rides")
plt.xlabel("Station Name")
```

```python
#End Stations graph
ends_only_df = final_df.loc[:, ["ride_id", "end_station_name"]]
end_stations = ends_only_df.groupby("end_station_name").count()
ends_sorted = end_stations.sort_values(by='ride_id', ascending=False)
top_50_ends = ends_sorted.head(50)
# top_50_ends
```

```python
top_50_ends.plot(kind="bar", figsize=(20,3), color="red")
plt.title("Usage of End Stations")
plt.ylabel("Number of rides")
plt.xlabel("Station Name")
```
