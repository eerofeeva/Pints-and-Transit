```python
#Find the top 50 stations
station_count = final_df['start_station_name'].value_counts()
```

```python
station_df = final_df.loc[:, ["ride_id", "start_station_name", "end_station_name"]]
station_latlng = final_df.loc[:, ["ride_id", "start_station_name", "end_station_name","start_lat", "start_lng", "end_lat", "end_lng"]]
starts_only_df = final_df.loc[:, ["ride_id", "start_station_name"]]
```

```python
stations_grouped = starts_only_df.groupby("start_station_name").count()
stations_grouped_df= stations_grouped.rename(columns={"start_station_name": "Station Name"})
stations_low=stations_grouped_df.sort_values(by="ride_id", ascending=True)
top_50 = stations_grouped_df.sort_values(by="ride_id", ascending=False).head(50)
# top_50
low_50 = stations_low.sort_values(by="ride_id", ascending=True).head(50)
# low_50
```

```python
#Bar graph of top 50

top_50.plot(kind="bar", figsize=(20,3))
plt.title("Most Popular Bike Stations")
plt.ylabel("Number of rides")
plt.xlabel("Station Name")

# plt.savefig("./graph images/top_50.png")
```

```python
#Bar graph of lowest 50 stations

low_50.plot(kind="bar", figsize=(20,3), color="red")
plt.ylim(0, 1000)
plt.title("Least Popular Bike Stations")
plt.ylabel("Number of rides")
plt.xlabel("Station Name")

# plt.savefig("./graph images/low_50.png")
```
